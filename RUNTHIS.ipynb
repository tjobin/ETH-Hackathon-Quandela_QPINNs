### So there is a difference between the quantum and classical models? But how much of this difference is due to the "quantumness" of the central layer?
To investigate this we tried looking at the "quantumness" of the latent space at the exit of the quantum layer. Although there is no universal way to do this, we looked at both the entanglement entropy and the Shannon entropy (of the superposition branches of the pure state output of the layer) to attempt to quantify how entangled and how much of in a superposition this output is.

```python
#Due to branch conflicts the code can not be run here, please look at the figures in the /entropies/ folder. Ignore the title on the figures themselves.
```

For both entropies, there is some transition hinting that quantum superposition and entanglement are more called upon at higher frequencies precisely where the quantum model performs best. These results however need more investigation
Also note that since those are classically simulated results they are no proof of quantum advantage, scaling at non-simmulable scales needs to be possible for this to represent a real advantage of QC.
